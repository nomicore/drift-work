```python
import uuid

def generate_id(row):
    # Creating a simple unique ID using 'TYRE' + hex part of uuid4
    return f"TYRE_{uuid.uuid4().hex[:6].upper()}"

# Remove commas from the "Price" column and convert to float
df['Price'] = df['Price'].astype(str).str.replace(',', '').astype(float)

df['id'] = df.apply(generate_id, axis=1)
df.head()
```

```python
# Save the DataFrame (including the generated 'id' column) to a JSON file
df.to_json('tyre_dataset_with_id.json', orient='records', lines=False, force_ascii=False)
```
